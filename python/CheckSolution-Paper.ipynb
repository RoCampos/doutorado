```python
import networkx as nx
import re

def check (edges, sources, members):
    G = nx.Graph ()
    G.add_edges_from (edges)
    print ("Tree: %s" % nx.is_tree (G))
    if not nx.is_tree (G):
        print ('Trees:',nx.number_connected_components (G))
        for g in nx.connected_component_subgraphs (G):
            print ("\tGraph :",nx.is_tree(g))
    print ("Connect: %s" % all (x in G.nodes () for x in members))
    group = sources + members
    lista = [x for x in G.nodes () if G.degree(x) == 1 and x not in group]
    print ("non-leaf(degre=1): %s(%s)" % (all(x in lista for x in group), len(lista)))
    if all(x in lista for x in group):
        for x in lista:
            print (x, G.degree(x), list(G.neighbors(x)))
    print ()
    
def check2 (edges, sources, members):
    print (edges)
    print (sources, members)
    print ()
    

with open('saida.txt') as f:
    sources = []
    edges = []
    members = []
    for line in f:
        if "SOURCE" in line:
            sources = list(map(int, re.findall (r'\d+',line)))
        elif "members" in line:
            members = list(map(int, re.findall (r'\d+',line)))
        elif "END" in line:
            check (edges, sources, members)
#             check2 (edges, sources, members)
            edges = []
            sources = []
            members = []
        else:
            e = list(map(int, re.findall(r'\d+', line)))
            if len(e) >= 2:
                edges.append (e[:2])
```
